BUILD THE MODEL

```python
import tensorflow as tf
from keras import layers
```

```python
# CNN model building

# create the input layer and tells the model to input shapes
input_layer = tf.keras.layers.Input(shape=(28,28,1))

# start building the model layers
model = tf.keras.models.Sequential([
    input_layer,
    layers.Conv2D(filters=10, kernel_size = (3,3), activation='relu'),#convulotional layer # Conv2d is used for the read image data or we say works with images , filters are like neurons that capture and store the information of the picture like there shape their colour, texture , height and width and so on , and this kernal is works like a window size it works like a frame that tells the model in which length and width you see the image , so kernal)size is works as a frame of the images and the filter is like a scanner of the images
    layers.Conv2D(filters=10, kernel_size = (3,3), activation='relu'),#convulotional layer # convulotional means apply a filter to extract the features from the data
    layers.MaxPooling2D(),
    layers.Conv2D(filters = 10, kernel_size = (3,3), activation = 'relu'), # pooling layer # it reduce the size of the image and just keep the important information from the image for faster training the model and avoids the risk of overfitting
    layers.Conv2D(filters = 10, kernel_size = (3,3), activation='relu'), # convolutional layer
    layers.Conv2D(filters = 10, kernel_size = (3,3), activation='relu'), # convolutionlal layer to learn from the image data
    layers.MaxPooling2D(), # pooling layer to normalize the data and keep the important features from the data and avoid to the risk of overfitting
    layers.Flatten(), # it can convert 2d, 3d, 4d image into the 1d image because out Dense layer(fully conected neurol layer) expects the data into the single vector so this Flatten layer merge the every image height , width and the colour channels of every image coming in the batch for example image is (28,28, 125) it merge everything together to become this images (4568)
    layers.Dense(10, activation='softmax') # this is the layer which gives the output, use 10 number of neurons of this fully connected layer(Dense) use their softmax function to finish the linear behavieor of the model take 10 neurons and use the oftmax because our task is multiclassification of 10 numbers each number of each neuron and use softmax for multiclass classification

])


```

```python
# Compilre the model
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Start training Model
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1 )
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
```

```python
def recognize_image(img):
    image = cv.imread(img,  cv.IMREAD_GRAYSCALE)
    gray = cv.bitwise_not(image)
    resize_image = cv.resize(gray, (28,28))
    prediction = model.predict(resize_image.reshape(1,28,28,1))
    num =  prediction.argmax()
    # print("The number in this image is {num}")
    print(f"the predicted number of model is {num}")
    return num

```
